# Creating a 5 letter Corpus for Wordle 

- We use the 10,000 word corpus by MIT (copied in wordlist.txt) and the European Parliment Data Token Corpus
- We take out the five letter words from them and remove names appearing in NLTK Names Corpus
- Then we manually go through the corpus to clean words which are not very common in english or obscene in nature
- The final cleaned corpus is used in the wordle game file

## Creating Corpus

```python
from tqdm.notebook import tqdm
```

```python
mit_wordlist = open("data/wordlist.txt", "r")
```

```python
word_corpus = []
for line in mit_wordlist.readlines():
    word_corpus.append(line.strip())
    
```

```python
len(word_corpus)
```

```python
word_corpus
```

### Removing Names from this

```python
female_names = open("data/names/female.txt", "r")
female_names_corpus = []
for line in female_names.readlines():
    female_names_corpus.append(line.strip())
    
```

```python
male_names = open("data/names/male.txt", "r")
male_names_corpus = []
for line in male_names.readlines():
    male_names_corpus.append(line.strip())
    
```

```python
female_names_corpus = [word.lower() for word in female_names_corpus]
```

```python
male_names_corpus = [word.lower() for word in male_names_corpus]
```

```python
word_corpus = set(word_corpus).difference(set(female_names_corpus).union(male_names_corpus))
```

```python
word_corpus.add("drown")
```

```python
word_corpus.add("trace")
```

```python
word_corpus.add("palms")
```

```python
word_corpus.add("tally")
```

### Checking for Punctuations

```python
puncts = ['\u200b','\xad','´','¡','˚','℅','«','÷','க',' ื','‟','‐',',', '.', '"', ':', ')', '(', '-', '!', '?', '|', ';', "'", '$', '&', '/', '[', ']', '>', '%', '=', '#', '*', '+', '\\', '•',  '~', '@', '£',
 '·', '_', '{', '}', '©', '^', '®', '`',  '<', '→', '°', '€', '™', '›',  '♥', '←', '×', '§', '″', '′', 'Â', '█', '½', 'à', '…',
 '“', '★', '”', '–', '●', 'â', '►', '−', '¢', '²', '¬', '░', '¶', '↑', '±', '¿', '▾', '═', '¦', '║', '―', '¥', '▓', '—', '‹', '─',
 '▒', '：', '¼', '⊕', '▼', '▪', '†', '■', '’', '▀', '¨', '▄', '♫', '☆', 'é', '¯', '♦', '¤', '▲', 'è', '¸', '¾', 'Ã', '⋅', '‘', '∞',
 '∙', '）', '↓', '、', '│', '（', '»', '，', '♪', '╩', '╚', '³', '・', '╦', '╣', '╔', '╗', '▬', '❤', 'ï', 'Ø', '¹', '≤', '‡', '√', '₹']

check_puncts = []

for word in tqdm(word_corpus):
    for p in puncts:
        if(p in word):
            check_puncts.append(word)
```

```python
check_puncts
```

```python
five_letter_eu = open("data/five_letter_corpus_eu.txt", "r")
five_letter_eu_corpus = set()
for line in five_letter_eu.readlines():
    five_letter_eu_corpus.add(line.strip())
```

```python
potential_names = five_letter_eu_corpus.intersection(set(female_names_corpus).union(male_names_corpus))
```

```python
len(five_letter_eu_corpus)
```

```python
for name in potential_names:
    print(name)
    to_remove = int(input("Want to remove this?"))
    if(to_remove==1):
        print("removing this")
        five_letter_eu_corpus.remove(name)
        print(len(five_letter_eu_corpus))
```

```python
check_puncts = []

for word in tqdm(five_letter_eu_corpus):
    for p in puncts:
        if(p in word):
            check_puncts.append(word)
```

```python
check_puncts
```

```python
five_letter_word_corpus = [word for word in word_corpus if len(word)==5]
```

```python
len(five_letter_word_corpus)
```

```python
len(five_letter_eu_corpus.union(five_letter_word_corpus))
```

```python
FIVE_LETTER_CORPUS = five_letter_eu_corpus.union(five_letter_word_corpus)
```

```python
final_corpus = open("FIVE_LETTER_CORPUS.txt", "w")
for word in FIVE_LETTER_CORPUS:
    final_corpus.write(word)
    final_corpus.write("\n")
    
final_corpus.close()
```

```python
CLEANED_FIVE_LETTER_CORPUS = set()
cleaned_corpus = open("FIVE_LETTER_CORPUS.txt", "r")
for line in cleaned_corpus.readlines():
    CLEANED_FIVE_LETTER_CORPUS.add(line.strip())
cleaned_corpus.close()

CLEANED_FIVE_LETTER_CORPUS.remove("")
```

```python
len(CLEANED_FIVE_LETTER_CORPUS)
```

```python
final_cleaned_corpus = open("WORDLE_FIVE_LETTER_CORPUS.txt", "w")
for word in CLEANED_FIVE_LETTER_CORPUS:
    final_cleaned_corpus.write(word)
    final_cleaned_corpus.write("\n")
    
final_cleaned_corpus.close()
```
